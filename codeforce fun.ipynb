```python
n = int(input())
res = ""
while n>0:
    rem = n%2
    res = str(rem)+res
    n = n // 2
print(res)
```

```python
#  reverse a string without usingg slicing
string = input()
rev = ""
for char in string:
    rev = char + rev
print(rev)
```

```python
def shift_zeros(arr,n):
    result = []
    for num in arr:
        if num != 0:
            result.append(num)
    zeros_count = n - len(result)
    result.append([0]*zeros_count)
    return result
n = int(input())
arr = list(map(int,input().split()))
ans = shift_zeros(arr , n)
print(*ans)
```

```python
def shift_zeros(arr, n):
    non_zeros = [x for x in arr if x!= 0]
    return non_zeros + [0] * (n - len(non_zeros))

n = int(input())
arr = list(map(int, input().split()))
print(*shift_zeros(arr, n))
```
