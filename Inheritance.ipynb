```python
#### Multi level inheritance

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def details(self):
        return f"Name: {self.name}, Age: {self.age}"


class Student(Person):
    def __init__(self, name, age, branch):
        super().__init__(name, age)
        self.branch = branch


class CollegeStudent(Student):
    def __init__(self, name, age, branch, college):
        super().__init__(name, age, branch)
        self.college = college


s = CollegeStudent("Saranya", 20, "CSE", "Brilliant College")

print(s.name)
print(s.age)
print(s.branch)
print(s.college)
print(s.details())
```

```python
 ### MULTIPLE

class Father:
    def father_details(self):
        return "Father: Ramu"


class Mother:
    def mother_details(self):
        return "Mother: Suma latha"


class Child(Father, Mother):
    def __init__(self, name):
        self.name = name


c = Child("Saranya")

print(c.name)
print(c.father_details())
print(c.mother_details())
```

```python
### Hierachical


class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age


class Student(Person):
    def student_info(self):
        return f"Student Name: {self.name}"


class Teacher(Person):
    def teacher_info(self):
        return f"Teacher Name: {self.name}"


s = Student("Saranya", 20)
t = Teacher("Rajesh", 35)

print(s.student_info())
print(t.teacher_info())
```
