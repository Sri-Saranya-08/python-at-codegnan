```python
#Tuple Operations
t = (10,20,30,40)
print(t)
```

```python
t = (10,20,30,40)
print(t[0])
print(t[2])
```

```python
t = (10,20,30,40)
print(len(t))
```

```python
#Tuple concatenation
t1 = (1,2,3)
t2 = (4,5,6)
result = t1+t2
print(result)
```

```python
t1 = (10,20,30)
t2 = (40,50,60)
result =t1+t2
print("Tuple 1:",t1)
print("Tuple 2:",t2)
print("Concatenated Tuple:",result)
```

```python
#Repeated Tuple
t = (1,2,3)
print(t*3)
```

```python
t = (10,20,30)
print(t*4)
```

```python
#Tuple Methods
#Tuple index()
t = (10,20,30,40,50)
print("Index of 30:",t.index(30))
```

```python
#Tuple Count()
t = (10,20,30,40,50)
print(t.count(20))
```
