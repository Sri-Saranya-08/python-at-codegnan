```python
#indexing
s="codegnan"
print(s[2])
print(s[-1])
```

```python
#slicing
#str_obj[start:end:step]
s="Codegnan"
print(s[2:7])
print(s[-4:])
print(s[5:2])
print(s[3:100])
print(s[::-1])
print(s[-3::-1])
print(s[-3::1])
```

```python
#empty string representation
#type() : this function returns the object datatype
s1=""
s2=str()
print(type(s1))
print(type(s2))
```

```python
#string reading from user
#input():this function is used to read dynamically from user and this function by default read it as string data type
s=input()
print(type(s))
print(s)
```

```python
#string operations
#1.string concatenation(using + operator)
#2.string repetation(using * operator)
s1="code"
s2="gnan"
s3=s1+" "+s2
print(s3)
print(s1+s1+s1)
print(s1 * 10)
```
