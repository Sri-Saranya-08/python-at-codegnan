```python
# Types of functions based on parameters and return value
#4. with parameters and with return

#function defination block
def addition(x:int,y:int)->int:
    c = x + y
    return c
#function calling block
a,b = 10,20
res = addition(a,b)
print(res)
```

```python
#3. with parameter and without return
#by default function return none value,when function not rreturning any value
#function definition bllock
def addition(x:int,y:int)->int:
    c = x + y
    print(f"This is from  function  definition block:{c}")
#function calling block
a,b = 10,20
res = addition(a,b)
print(f"This is from function calling block:{res}")
```

```python
#2.without parameters and with return
#function definition bllock
def addition()->int:
    x = 10
    y = 20
    z = x + y
    return z

#function calling block

res = addition()
print(f"This is from function calling block:{res}")
```

```python
#even and odd using functions
def even_or_odd(x:int)->str:
    if x % 2 == 0:
        return "Even"
    else:
        return "odd"
#function calling block
n = int(input())
print(even_or_odd(n))
```

```python
#prime numbers using function
def check_prime (n:int)->str:
    if n < 2:
        return "Not prime"
    for i in range (2, n//2+1):
        if n % i == 0:
            return "Not prime"
    return "prime"
#function calling
num = int(input())
print(check_prime(num))
    
```
