```python
#Scope of variable
#1.Local scope
#2.enclosing scope
#3.global scope
#4.buillt in scope

x = 10 #Global scope variable
def outer():
    y = 20 #Enclosing scope variable
    def inner():
        z = 30 #Local scope variable
        print(z, y, x) #print() - Built in scope function
    inner()
outer()
```

```python
#using global scope variable in local scope
x = 100
def  outer():
    y = x + 1
    print("Local scope variable value:",y)
outer()
print("Global scope variable value:",x)
```

```python
#updating the global scope variable value in local scope
#can't update global scope variable in local scope directly
x = 100
def  outer():
    x = x + 1
    print("Local scope variable value:",x)
outer()
print("Global scope variable value:",x)
```

```python
#updating the global scope variable value in local scope by using 'global' keyword
#
x = 100
def  outer():
    global x
    x = x + 1
    print("Local scope variable value:",x)
outer()
print("Global scope variable value:",x)
```
