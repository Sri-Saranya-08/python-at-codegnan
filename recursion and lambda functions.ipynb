```python
#print 1 to n using recursion
def fun(x):
    if x == 0:
        return 0
    fun(x-1)
    print(x)
fun(3)
```

```python
def fun(x):
    if x == 0:
        return 0
    print(x)
    fun(x - 1)
    
fun (3) 
```

```python
#Factorial
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)
print(factorial(5))
```

```python
#sum of digits in a number
# n = 456 o/p : 15
def sumofdigits(n):
    if n == 0:
        return 0
    return n % 10 + sumofdigits(n//10)
print(sumofdigits(456))
print(sumofdigits(15346))
```

```python
#count digits in a number
# n = 456 o/p : 15
def sumofdigits(n):
    if n == 0:
        return 0
    return 1 + sumofdigits(n//10)
print(sumofdigits(456))
print(sumofdigits(15346))
```

```python
# find the square of a number using lambda
s = lambda x:x**2
print(s(10))
```

```python
# find the sqyare of list of integer using map and lambda
arr = [1,2,3,4,5]
print(list(map(lambda x:x**2 , arr)))
```

```python
# find the frequency of eachh character
string = input()
count = {}
for char in string:
    if char in count:
        count[char] += 1
    else:
        count[char] = 1
print(count)
```
