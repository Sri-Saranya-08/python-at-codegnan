```python
#find the differennce between two strings
s1 = "CODE"
s2 = "GNAN"
n = len(s1)
s = ""
for i in range (n):
    diff = abs(ord(s1[i]) - ord(s2[i]))
    if diff == 0:
        s += s1[i]
    else:
        s += chr(64+diff)
print(s)
            
```

```python
#find bigger and smaller character from two strings
#compare each character in both strings abd if s1[i]>s2[i] add [s1[i] to bigger and s2[i]to
#if both same levell it
s1 , s2 = map(str,input().upper . split())
n = len (s1)
big = small = ""
for i in range(n):
    if s1[i] !=s2[i]:
        if s1[i] > s2[i]:
            big += s1[i]
            small += s2[i]
        else:
            big += s2[i]
            small += s1[i]
print(big)
print(small)
    
```

```python
# map(fun,iterable):It applies the functionality on each element in iterable
# filter(fun,iterable)"It filteers the iterables,it returns the vallues which is not none
# FInd the square of the list of integers
def square(x):
    return x**2
lst = [1,2,3,4,5]
print(list(map(square,lst)))
```

```python
#filter the even numbers from list off tge integers
arr=[1,2,3,4,5,6,7]
def even(x):
    if x%2 == 0:
        return 1
arr = [1,2,3,4,5,6,7]
print(list(filter(even,arr)))
print(list(map(even,arr)))
```

```python
def square(t):
    x,y = t
    return x+y
lst = [(1,2),(3,4)]
print(list(map(square,lst)))
```

```python
#Find the target element possition in list
arr = [10,20,30,40,50]
target = 400
for ind , val in enumerate(arr):
    if target == val:
        print(ind)
        break
else:
    print(-1)
```

```python
# any(iterable):It returns True when any one element's boolean value is true
arr = [1,2,4,0,0,0]
print(any(arr))
print(any([0,0,0,0]))
# all(iterable):It returns True when all element's boolean value is true
print(all([1,2,3,4]))
print(all([True,False,True]))
print(all([False,False,False]))
```
