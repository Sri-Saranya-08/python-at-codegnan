```python
# define person class
class Student:
    school = "XYZ school" # class attributes
    def __init__(self, name, age): # Initialization Constructot (Magic method)
        self.name = name # Instance attribute
        self.age = age # Instance attribute
        #print("This Student class __init__ method")
    def details(self):
        return f"My name is {self.name} and i am  {self.age} years old"

s1 = Student("Ravi", 20)
s2 = Student("janu", 19)
print(s1)
print(s1.name)
print(s2.age)
print(s1.details())
print(s1.school)
print(s2.school)
```
