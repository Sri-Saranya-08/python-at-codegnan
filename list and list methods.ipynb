```python
###List Methods
-Item adding methods
 -append(item):It adds item at end of list
 -extend(iterator):It adds the all itreator elements at end of list 
 -insert(pos,item):It adds item at specified position
```

```python
lst = [10,20,30,40]
lst.append(100)
print(lst)
lst.extend([1,2,3])
print(lst)
lst.append([100,200])
print(lst)
lst.insert(0,1000)
print(lst)
```

```python
###Item Removing Methods
   -remove(item):It Removes the first occurances item form the list if it is exists in
list otherwise 'ValueError
   -pop(pos):It removes and return the specified position item if position is passed,
otherwise it remove and returns the last item from last
   -clear():it removes all elements ffrom list
```

```python
lst = [10,20,30,20,40,50,20]
lst.remove(20)
print(lst)
#lst.remove(100)
print(lst)
```

```python
lst = [10,20,30,20,40,50,20]
val = lst.pop()
print(val)
print(lst)
val2 = lst.pop(0)#it removes the 0th position
print(val2)
print(lst)
lst.clear()
print(lst)
```

```python
#index and count()
#index(item):If item is present in list it returns the index othrewise
#                    valueerror
#count(item):It returns the count of total occurances of item
lst = [10,20,30,20,40,50]
ind = lst.index(20)
print(ind)
print(lst.count(20))
```

```python
#Sorting items
#sort(reverse=False):By default it stores the origin a list in into ascending order,
#        if reverser is true it sorts in decending order
lst = [1,4,2,5,-5,3,-45,-4,30]
lst.sort()
print(lst)
lst.sort(reverse = True)
print(lst)
```
