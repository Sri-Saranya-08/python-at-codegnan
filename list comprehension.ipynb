```python
# in given arr of integers perform following operations
# if-even - find square of a number
# if-odd - find squareroot value (upto 2 decimal places)
arr = list(map(int,input().split()))
res = [x**2 if x%2==0 else round(x**0.5,2) for x in arr]
print(res)
```

```python
# find out 2 dices combiination values
res = [(x,y) for x in range(1,7)for y in range (1,7)]
print(res)
```

```python
# return the dice combination if sum of two face values is divisible by 3
res = [(x,y) for x in range(1,7)for y in range(1,7) if (x+y)%3==0]
print(res)
```

```python
# set comprehension
# 1 to n numbers
n = int(input())
res = {x for x in range(1,n+1)}
print(type(res))
print(res)
```

```python
# dictionary comprehension
# 1 to n numbers
n = int(input())
res = {x:x**2 for x in range(1,n+1)}
print(type(res))
print(res)
```

```python
# tuple comprehension
# 1 to n numbers
n = int(input())
res =tuple (x  for x in range(1,n+1))
print(type(res))
print(res)
```
