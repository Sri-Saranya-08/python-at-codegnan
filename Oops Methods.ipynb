- Static Method: The method defination starts eith `staticmathod` decorator and it doesn't take first parameter as self or cls those methods called as `static mathod` .This methods works on except instance attributes and class attributes.`syntax`:
          
              `        class class_name:`

                            #block code
                         

```python
# class ttribute - school
# Instance attributes - id,name class

class Student:
    school = "Codegnan School" # class attribute
    def __init__(self,stdid,name,standard):
        self.name = name # Instance attribute
        self.stdid = stdid # Instance attribute
        self.standard = standard # Instance attribute
    def student_details(self): #  Instance Methods
        return f"My Self {self.name} and i am from {self.standard} class"
    # class Method
    @classmethod
    def update_school(cls,new_school_name):
        cls.school = new_school_name
        return "school name updated"
    @staticmethod
    def percentage(marks:list):
        return (sum(marks)/300)*100
s1 = Student(101, "saranya" ,9)
s2 = Student(102, "rajesh" ,8)
print(s1.school , s2.school)
print(s1.student_details())
print(s1.update_school("CODEGNAN SCHOOL"))
print(s1.school,s2.school)
print(s2.percentage([85,68,95]))

    
```

#### Encapsulation:
  ` - Encapsulation means,wraping the methods and attributes in single unit that one is class. 
The maain aim of encapsulation is to restrict tthe direct access of attributes and methods.
    - This type of restriction can implement by using `Access Modifiers`
 `    - types of access modifiers
    1. public
    2. protected
    3. private                                                                                         
`- In python we don't have any keywords to represent it public,protected and private,but by using underscore(_) can define`
`   1. public : The public Members(i.e attributes + method) does't starts with underscore.This public members can access any
                where in the program.`
`   2. protected : The protected Members(i.e attributes + method) start with single underscore(_) .The members can access in inside the class abd in subclass also.
   3. private :  The private Members(i.e attributes + method) start with double underscore(__) .The members can access in only inside`
    - python doesn't and support high security like java.here we can access the protected members


    - here to access thr protected members :obj (protected member
    - to access thr private members by using name handlimg
    synyax: obj._classname__private_member_name
    -the advantage of encapsulation  is can provide the security thhe dataas well as and functionality
